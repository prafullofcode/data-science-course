1. Matrix Operation
Perform the following operations on the given matrices:
A=[[2    -1], [​0    3​]]
B=[[1   ​ 2], [-3      4​]]
a. A+B
b. AB
c. 3A−2B

```python
A = np.array([
        [2,-1],
        [0,3]
    ])
B = np.array([
    [1,2],
    [-3,4]
])
addition_of_matrices = A+B
product = A @ B
operation = 3*A - 2*B

print(addition_of_matrices)
print(product)
print(operation)
```

2. Determinants and Inverses  
    Find the determinant of A.
    Determine whether A is invertible. If yes, calculate A^−1.

```python
determinant = np.linalg.det(A)
print(determinant)
if not determinant.round() == 0:
    print("Inverse is possible as A is non singular")
    inverse_of_A = np.linalg.inv(A)
print(inverse_of_A)
```

Eigenvalues and Eigenvectors  

    Compute the eigenvalues and eigenvectors of the matrix:
    C=[[4    1], [​−2       1]​]

4. 

```python
C = np.array([
    [4,1],
    [-2,1]
])

eigenvalues, eigenvectors = np.linalg.eig(C)
print("Eigen values are:",eigenvalues)
print("Eigen vectors are:",eigenvectors)
```

```python
X = np.array([
    [4, 1.5, 100],   
     [6, 2.0, 150]   
])

Y = np.array([20, 30])


beta = np.linalg.pinv(X.T @ X) @ X.T @ Y

# beta = np.linalg.pinv(X) @ Y

# beta = np.linalg.lstsq(X, Y,rcond=None)[0] #<- returns a tuple with first elements solution , then other stuffs too
print("β =", beta)

car3= np.array([
    [8, 3.0, 200]
])

price_car3 = car3 @ beta

#dot product returns a matrix so [0]
print(f" price of car 3 : {price_car3[0]}")
```
