### Visualization

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='blue', marker='+', linestyle='dotted')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axisYears
plt.ylabel("Billions of $")
plt.xlabel("Years")
plt.show()

```

```python
name = ['Ram' , 'Shyam' ,'Hari']
age = [12,13,15]

plt.plot(name,age,color = 'green',marker = '*')

plt.title("Practice")

plt.ylabel("Age")
plt.xlabel("Name")

plt.show()
```
