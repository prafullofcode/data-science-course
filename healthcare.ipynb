### Data Visualization Using a Healthcare Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Dataset/Health_heart_experimental.csv")
df.describe()
```

### Line Plot

```python
sns.lineplot(x='BMI', y='heightCm', hue = 'indication' ,data=df)
```

Indicates a inverse relationship between BMI and height
The graph shows that individuals with low BMI who are tall (> ~170cm), as well as those with very high BMI and shorter height have poor heart health(indication = 1). 


### Bar Plot

```python
sns.barplot(x='sex', y='BMI', hue = 'indication', data=df)
```

The graph shows that individuals the average BMI of both male and female , irrespective of heart health indication is almost the same. Which indicates that heart health is not influenced by gender but more so by BMI.

### Pie chart

```python

df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

The pie chart indicates even distribution of male and female in the collected data

### Box Plot

```python
sns.boxplot(x='sex', y='HR', hue='indication', data=df)
```

The median for healthy heart males and females is almost the same and same goes for ones who are unhealthy(indication = 1). Further supporting the argument that heart health is not influenced by gender.

### Scatter Plot

```python
sns.scatterplot(x='age', y='BMI', data=df, hue='indication')
plt.show()
```

The scatter plot indicates that regardless of age, the determining factor for heart health indication is BMI. Since the patients with lower BMI all have unhealthy heart and same goes for the ones with extremely high BMI. 

```python
sns.scatterplot(x='age', y='BMI', data=df, hue='indication')
plt.ylim(14, 20)
plt.show()
sns.scatterplot(x='age', y='BMI', data=df, hue='indication')
plt.ylim(29,34)
plt.show()
```

Zooming in on specific values of BMI we can clearly observe that the population with BMI > 31 is surely suffering from heart related health issues and the population with BMI < 17 is also suffering from heart related health issue.
